```python
#CREATE MODEL
#DEPENDENT AND INDEPENDENT FEATURES
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from os import stat
#Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
#standadScaller
#global minima~~~ different units~~ converging fast to global M.
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_test
```

model training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
#print Coeff and intercepts
regression.coef_
```

```python
regression.intercept_
```

```python
#on which the model trained
regression.get_params()
```

```python
#prediction with test data
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

# Assumptions

```python
#plot a scatter plot for the prediction, True value vs prediction
plt.scatter(y_test,reg_pred)
```

```python
#prediction with residuals,,,error
residuals=y_test-reg_pred
```

```python
residuals
```

```python
#plotting resudials
sns.displot(residuals,kind='kde')
```

```python
#scatter plot w respect to predictions and residuals.....no any kind of distribution,
#scattered uniformily
plt.scatter(reg_pred,residuals)
```

```python
#model performance
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

```python
#R square and Adjusted R square
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

```python
#Adjusted r2,use only formula and it is always less than r2.
```
